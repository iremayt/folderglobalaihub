```python
import pandas as pd
import numpy as np
```

```python
k1 = pd.DataFrame({
    "studentName":["",""],
    "midtermGrade":["",""],
    "projectGrade":["",""],
    "finalGrade":["",""],
    "passingGrade":["",""],

})

def check1(a):
    while True:
        try:
            if a >= 0 and a <= 100:
                return a
            
            elif a<0 or a>100:
                print("0-100 arasi gir!!")
                check1(a)
                
            elif a == str(a):
                print("harf girdin dÃ¼zelt!!")

        except:
            print("integer sayi gir!!")
```

```python
#get students name
for i in range(2):
    print(f"\n Enter for student{i+1}: ")
    k1.studentName[i] = str(input("Name:   "))
    
    #get grades
    k1.midtermGrade[i] = int(input("Midterm Grade:   "))
    check1(k1.midtermGrade[i])
    k1.projectGrade[i] = int(input("Project Grade:   "))
    k1.finalGrade[i] = int(input("Final Grade:     "))
    
    #calculate passing grade
    k1.passingGrade[i] = (k1.midtermGrade[i]*0.3 + k1.projectGrade[i]*0.3 + k1.finalGrade[i]*0.4)
    
#sort the table according to passing grades
yenik1 = k1.sort_values('passingGrade', ascending=False)

#print the sorted table
print(yenik1)
```
